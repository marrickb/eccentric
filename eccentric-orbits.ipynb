```python
import numpy as np

def anomalies(t, t_0, P, e):
    """
    Calculate the mean anomaly and true anomaly for an orbit.

    Parameters:
    t (float): Current time.
    t_0 (float): Reference time (e.g., periapsis passage).
    P (float): Orbital period.
    e (float): Orbital eccentricity.

    Returns:
    tuple: Mean anomaly (degrees), True anomaly (degrees).
    """
    # Mean anomaly in radians
    m_a_rad = 2 * np.pi * (t - t_0) / P % (2 * np.pi)

    # True anomaly approximation (radians)
    v_a_rad = (m_a_rad +
               (2 * e - (e**3) / 4) * np.sin(m_a_rad) +
               (5 / 4) * (e**2) * np.sin(2 * m_a_rad) +
               (13 / 12) * (e**3) * np.sin(3 * m_a_rad)) % (2 * np.pi)
    print(v_a_rad)
    # Convert both anomalies to degrees for output
    m_a_deg = np.degrees(m_a_rad)
    v_a_deg = np.degrees(v_a_rad)

    print(f"Mean anomaly: {m_a_deg:.2f} deg")
    print(f"True anomaly: {v_a_deg:.2f} deg")
    
    return m_a_deg, v_a_deg

# Example usage
# anomalies(t=10, t_0=0, P=365.25, e=0.0167)
```

```python
def eccentric_orbits(e=0.3,L_starlog=(0), R_star=1, sma=(1*149597870700), t_slice=0,t_0=3.1985745555555556,t_sim=17000,P=11.186):
    max_toa_sw = {}
    # for cube in cubes:
    #     if cube.standard_name in ['toa_incoming_shortwave_flux']: #, 'surface_temperature'
    #         # Extract the time slice and copy it
    #         data_slice = cube[t_slice, 0, :, :].copy()
            
    #         # Find the maximum value of TOA SW flux and its indices
    #         max_value = data_slice.data.max()
    #         max_index = (data_slice.data == max_value).nonzero()

    #         # Extract the corresponding latitude and longitude
    #         lat_coord = data_slice.coord('latitude')
    #         lon_coord = data_slice.coord('longitude')

    #         # Retrieve the exact grid points from indices
    #         latitude = lat_coord.points[max_index[0]]
    #         longitude = lon_coord.points[max_index[1]]

    #         # Store the coordinates of max TOA SW flux in dictionary
    #         max_toa_sw[cube.standard_name] = {
    #             'max_value': max_value,
    #             'latitude': latitude,
    #             'longitude': longitude
    #         }
            
    #         print(f"Current Time slice:: {t_slice} days")
            # first current time in Julian dates (See UM023 radiation paper and source code)
            # uses Julian dates for 00:00:00 01-01-0001, plus planetary epoch 2451545 (00:30:00 UT, 01-01-2013)
            # t_slice for start of cube contents, t_sim for spin-up time (17000 days for 3:2), t_0 for time to longitude of perihelion
        #day_nr_midnight = 1721425.5 + 365*(2000-1) + (2000-1)/4 - (2000-1)/100 + (2000-1)/400 + t_slice +t_sim-t_0-1.0 - 2451545
            # Convert to noon on current day (86400 seconds per day)
        #day_number = day_nr_midnight + (43200+240/2)/86400
            #print(day_nr_midnight)
    #print(f"The current day number: {t_slice} days")
            # Mean anomaly in radians
    t_slice = t_slice/24
    m_a_rad =  (6.2400214 + (2*np.pi/P)*(t_slice)) % (2*np.pi) # -np.pi (+t_0)
    #print(f"Mean anomaly: {np.degrees(m_a_rad)} deg")
            
            # gamph represents pi minus longitude of perihelion (radians)
    gamph = np.pi - 1.796601474

            # hour angle for equation of time later (radians)
    hour_angle = (0.0 + (np.pi/P)*(t_slice))# % (2 * np.pi)
            
            # True anomaly approximation (radians), from Smart (1944)
    v_a_rad = (m_a_rad +
               (2 * e - (e**3) / 4) * np.sin(m_a_rad) +
               (5 / 4) * (e**2) * np.sin(2 * m_a_rad) +
               (13 / 12) * (e**3) * np.sin(3 * m_a_rad))% (2 * np.pi)
    v_a_deg = np.degrees(v_a_rad % (2*np.pi))
    #print(f"True anomaly: {(v_a_deg)} deg") 
    R_star= R_star*(6.957e8) #m
    L_star = 10**(L_starlog)*(3.828*10**26) #W
    #print(L_star)
    S = L_star/(4*np.pi*(sma)**2)*(1/(1-e**2))**2*((1+e*np.cos(v_a_rad)))**2 #+e**2/2) #W/m2
            
    sindec = np.sin(0)*np.sin(v_a_rad-gamph)

        # Calculation of the equation of time for discrepancy between mean solar and clock time
        # following Mueller (Acta Phys. Polo. A 88 Supplement, S-49, 1995)
    epsilon = 0 # Obliquity, 0 here
    y = (np.tan(0.5*epsilon))**2 
    y2 = y*y
    e2 = e**2
    p = 0.5*np.pi - gamph
    eqt = (-y * (1.0-4.0*e2) * np.sin(2*(m_a_rad+p)) - 2.0*e*np.sin(m_a_rad) + 
                   2.0*e*y*np.sin(m_a_rad+2*p) - 2.0*e*y*np.sin(3.0*m_a_rad+2.0*p) -
                   0.5*y2*np.sin(4.0*(m_a_rad+p)) - 1.25*e2*np.sin(2.0*m_a_rad) +
                   2.0*e*y2*np.sin(3.0*m_a_rad+4.0*p) - 3.25*e2*y*np.sin(4.0*m_a_rad+2.0*p)
                   - y2*y*np.sin(6.0*(m_a_rad+p))/3.0)
            # correction for planets hour angle from eq of time
    eqt_f = eqt + (hour_angle) % (2*np.pi) # + np.pi
    #print(f"Corrected hour angle: {np.degrees(eqt_f)} deg") 
        #print(f"Hour angle: {(np.degrees(hour_angle)) %360} deg") 
        # Orbital phase angle as seen by distant observer
    phase_angle = (v_a_rad - gamph + np.pi + 0) % (2*np.pi)
    #print(f"Phase angle: {np.degrees(phase_angle)} deg") 
    #print(f'Calculated S: {S} W/m2')

        #print(f"Maximum {cube.standard_name.replace('_', ' ').title()}: {max_value} W/m2")
        #print(f"Grid point with maximum value - Latitude: {latitude}, Longitude: {longitude}")
    
    return S #np.degrees(phase_angle), (np.degrees(eqt_f) % 360), v_a_deg

# Example usage
# results = max_latlon(cubes, t_slice=0)
# print(results)
```

```python

def plot_flux_evol(L_starlog, R_star, sma, timerange, timefreq, P, title, save):
    time_flux_pcb_e0 = []
    time_flux_pcb_e01 = []
    time_flux_pcb_e02 = []
    time_flux_pcb_e03 = []
    time_flux_pcb_e04 = []
    for i in range(0,timerange,timefreq):
        S_e0 = eccentric_orbits(e=0, L_starlog=L_starlog, R_star=R_star, sma=sma, t_slice=i, P=P)
        S_e01 = eccentric_orbits(e=0.1, L_starlog=L_starlog, R_star=R_star, sma=sma, t_slice=i, P=P)
        S_e02 = eccentric_orbits(e=0.2, L_starlog=L_starlog, R_star=R_star, sma=sma, t_slice=i, P=P)
        S_e03 = eccentric_orbits(e=0.3, L_starlog=L_starlog, R_star=R_star, sma=sma, t_slice=i, P=P)
        S_e04 = eccentric_orbits(e=0.4, L_starlog=L_starlog, R_star=R_star, sma=sma, t_slice=i, P=P)
        time_flux_pcb_e0.append(S_e0)
        time_flux_pcb_e01.append(S_e01)
        time_flux_pcb_e02.append(S_e02)
        time_flux_pcb_e03.append(S_e03)
        time_flux_pcb_e04.append(S_e04)
    
    xpos=[]
    xlab=[]
    xpos_p=[]
    xlab_p=[]
    colors = plt.cm.viridis(np.linspace(0, 1, 5))
    for i in range(0,int(timerange/24*P)+1):
        xpos.append(0+24*P*i)
        xpos.append(24*P/2+24*P*i)
    for i in range(len(xpos)):
        if i % 2 == 0:
            xlab.append('peri')
        else:
            xlab.append('apo')  
    plt.figure(figsize=(9, 6))
    plt.plot(range(0,timerange,timefreq), time_flux_pcb_e0, color=colors[0], label='e=0.0')
    plt.plot(range(0,timerange,timefreq), time_flux_pcb_e01, color=colors[1], label='e=0.1')
    plt.axhline(np.mean(time_flux_pcb_e01), color=colors[1], ls='--')
    plt.plot(range(0,timerange,timefreq), time_flux_pcb_e02, color=colors[2], label='e=0.2')
    plt.axhline(np.mean(time_flux_pcb_e02), color=colors[2], ls='--')
    plt.plot(range(0,timerange,timefreq), time_flux_pcb_e03, color=colors[3], label='e=0.3')
    plt.axhline(np.mean(time_flux_pcb_e03), color=colors[3], ls='--')
    plt.plot(range(0,timerange,timefreq), time_flux_pcb_e04, color=colors[4], label='e=0.4')
    plt.axhline(np.mean(time_flux_pcb_e04), color=colors[4], ls='--', label='time-mean')
    plt.xlabel('Orbit', fontsize=17)
    plt.ylabel(r'S$_{\mathrm{TOA}}$ (Wm$^{-2}$)', fontsize=17)
    plt.xticks(xpos, labels=xlab, fontsize=16, rotation=75)
    plt.yticks(fontsize=11)
    plt.xlim(1, timerange)
    for i in range(0, int(timerange/24*P)+1):
    #     #plt.axvline(11.186 * i, ls='dashdot', color='grey')
        plt.axvline(xpos[i*2], ls='dotted', color='grey', alpha=0.3)
        plt.axvline(xpos[(i)*2-1], ls='dashdot', color='grey', alpha=0.3)
    plt.legend(loc='upper right', frameon=True, prop={'size': 16})#, bbox_to_anchor=(1.23, 0.5))
    if title:
        plt.title(f'{title}', fontsize=22)
    if save:
        plt.savefig(f'plots/orbit_{title}.png', bbox_inches='tight')
    plt.show()
    print(f'mean flux for e01: {np.mean(time_flux_pcb_e01)} W/m2')
plot_flux_evol(L_starlog=(-2.8), R_star=0.141, sma=(0.0485*149597870700), timerange=3600,timefreq=6, P=11.186, title='Proxima Centauri b', save=True)
plot_flux_evol(L_starlog=(-2.841), R_star=0.1556, sma=(0.03984*149597870700), timerange=2400,timefreq=6, P=8.46, title='LP 890-9 c - SPECULOOS-2 c', save=True)
plot_flux_evol(L_starlog=(-3.26), R_star=0.1192, sma=(0.02227*149597870700), timerange=1200,timefreq=6, P=4.049219, title='TRAPPIST-1 d', save=True)
plot_flux_evol(L_starlog=(-3.26), R_star=0.1192, sma=(0.02925*149597870700), timerange=1200,timefreq=6, P=6.101013, title='TRAPPIST-1 e', save=True)
plot_flux_evol(L_starlog=(-3.364516), R_star=0.1192, sma=(0.02227*149597870700), timerange=1200,timefreq=6, P=4.049219, title='Fake TRAPPIST-1 d', save=True)
plot_flux_evol(L_starlog=(-0.825593), R_star=0.618522, sma=(0.409*149597870700), timerange=12000,timefreq=24, P=112.3053, title='Kepler-442 b', save=True)

plot_flux_evol(L_starlog=(-1.64), R_star=0.421, sma=(0.1633*149597870700), timerange=3600,timefreq=24, P=37.42396, title='TOI-700 d', save=True)
plot_flux_evol(L_starlog=(-1.64), R_star=0.421, sma=(0.1340*149597870700), timerange=3600,timefreq=24, P=27.80978, title='TOI-700 e', save=True)
plot_flux_evol(L_starlog=(-3.14), R_star=0.120, sma=(0.0259*149597870700), timerange=1200,timefreq=6, P=4.90634, title='Teegarden b', save=True)

eccentric_orbits(e=0.0,L_starlog=(0), R_star=1, sma=(1*149597870700), t_slice=10,t_0=3.1985745555555556,t_sim=17000,P=8.46)
```
